# Visualization mode #

```python
from chip import *
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

target = chip()

# origin
target.offset_x = 0.0
target.offset_y = 0.0

# max value for X Y
target.x_max = 60.0
target.y_max = 60.0

# steps
target.steps = 10

target.lr = False
target.random = False
target.vertical = False
target.reverse = False

myiter = iter(target) # dummy iterator
XYpairs = target.coordinates

img = Image.open('nrf.png')
plt.rcParams['figure.figsize'] = [100,100]

ax = plt.axes()
for a in range(0,len(XYpairs)-1):
    ax.arrow(XYpairs[a][0],XYpairs[a][1], XYpairs[a+1][0]-XYpairs[a][0], XYpairs[a+1][1]-XYpairs[a][1], head_starts_at_zero=True,head_width=2, head_length=1, lw=5, ec='blue')
plt.imshow(np.flipud(img.resize((int(target.x_max+1), int(target.y_max+1)))), origin='lower')
plt.show()
```
