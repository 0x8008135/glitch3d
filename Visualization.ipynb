# Visualization mode #

```python
from chip import *
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
target = chip()

target.set_home(0.0,0.0)
target.set_end(15.0,15.0)

# steps
target.steps = 1
#target.spiral()
#positions = [p for p in target.spiral_inward()]
#fix outward offset out...
positions = [p for p in target.spiral_outward()]

#print(positions)
```

```python
img = Image.open('nrf.png')
plt.rcParams['figure.figsize'] = [100,100]

ax = plt.axes()
for a in range(0,len(positions)-1):
    ax.arrow(positions[a][0],positions[a][1], positions[a+1][0]-positions[a][0], positions[a+1][1]-positions[a][1], head_starts_at_zero=True, head_width=0.2, head_length=0.1, lw=5, ec='blue')
plt.imshow(np.flipud(img.resize((int(target._end_x+1), int(target._end_y+1)))), origin='lower')
plt.show()
```
